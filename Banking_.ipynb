```python
#Step 1: Load Libraries and Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/banking_data.csv'  # Adjust path if needed
data = pd.read_csv(file_path)

# Display first few rows
print(data.head())
```

```python
# Step 2: Analysis and Visualizations

#1. Distribution of Age Among Clients
plt.figure(figsize=(8, 5))
sns.histplot(data['age'], kde=True, color='blue', bins=20)
plt.title('Distribution of Age Among Clients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Variation in Job Types

plt.figure(figsize=(10, 6))
sns.countplot(y='job', data=data, order=data['job'].value_counts().index, palette='viridis')
plt.title('Variation in Job Types Among Clients')
plt.xlabel('Count')
plt.ylabel('Job Type')
plt.show()
```

```python
# 3. Marital Status Distribution

plt.figure(figsize=(6, 4))
data['marital'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
plt.title('Marital Status Distribution')
plt.ylabel('')
plt.show()
```

```python
# 4. Level of Education

plt.figure(figsize=(8, 5))
sns.countplot(x='education', data=data, order=data['education'].value_counts().index, palette='muted')
plt.title('Level of Education Among Clients')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.show()
```

```python
# 5. Proportion of Clients with Credit in Default

credit_default = data['default'].value_counts()
print(credit_default)

credit_default.plot(kind='bar', color=['green', 'red'])
plt.title('Proportion of Clients with Credit in Default')
plt.xlabel('Default')
plt.ylabel('Count')
plt.show()
```

```python
# 6. Distribution of Average Yearly Balance
plt.figure(figsize=(8, 5))
sns.histplot(data['balance'], kde=True, color='purple')
plt.title('Distribution of Average Yearly Balance')
plt.xlabel('Balance (in Euros)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 7. Clients with Housing Loans
housing_loans = data['housing'].value_counts()
print(housing_loans)

housing_loans.plot(kind='bar', color=['orange', 'blue'])
plt.title('Clients with Housing Loans')
plt.xlabel('Housing Loan')
plt.ylabel('Count')
plt.show()
```

```python
# 8. Clients with Personal Loans

personal_loans = data['loan'].value_counts()
print(personal_loans)

personal_loans.plot(kind='bar', color=['cyan', 'pink'])
plt.title('Clients with Personal Loans')
plt.xlabel('Personal Loan')
plt.ylabel('Count')
plt.show()
```

```python
#9. Communication Types Used
plt.figure(figsize=(6, 4))
sns.countplot(x='contact', data=data, order=data['contact'].value_counts().index, palette='cool')
plt.title('Communication Types Used During Campaign')
plt.xlabel('Contact Type')
plt.ylabel('Count')
plt.show()

```

```python
# 10. Distribution of Last Contact Day of the Month
plt.figure(figsize=(8, 5))
sns.histplot(data['day'], bins=31, color='teal')
plt.title('Distribution of Last Contact Day of the Month')
plt.xlabel('Day')
plt.ylabel('Frequency')
plt.show()
```

```python
# 11. Last Contact Month Variation
plt.figure(figsize=(8, 5))
sns.countplot(x='month', data=data, order=['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec'], palette='magma')
plt.title('Last Contact Month Variation')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

```python
# 12. Distribution of Last Contact Duration
plt.figure(figsize=(8, 5))
sns.histplot(data['duration'], kde=True, color='lime')
plt.title('Distribution of Last Contact Duration')
plt.xlabel('Duration (in Seconds)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 13. Contacts Performed During Campaign
plt.figure(figsize=(8, 5))
sns.histplot(data['campaign'], bins=20, color='coral', kde=False)
plt.title('Number of Contacts Performed During Campaign')
plt.xlabel('Number of Contacts')
plt.ylabel('Frequency')
plt.show()
```

```python
# 14. Days Passed Since Last Contact
plt.figure(figsize=(8, 5))
sns.histplot(data['pdays'], kde=False, color='brown', bins=50)
plt.title('Distribution of Days Passed Since Last Contact')
plt.xlabel('Days')
plt.ylabel('Frequency')
plt.show()
```

```python
# 15. Contacts Before Current Campaign
plt.figure(figsize=(8, 5))
sns.histplot(data['previous'], bins=15, color='darkblue', kde=False)
plt.title('Number of Contacts Before Current Campaign')
plt.xlabel('Number of Contacts')
plt.ylabel('Frequency')
plt.show()
```

```python
# 16. Outcomes of Previous Marketing Campaigns
plt.figure(figsize=(8, 5))
sns.countplot(x='poutcome', data=data, order=data['poutcome'].value_counts().index, palette='Set2')
plt.title('Outcomes of Previous Marketing Campaigns')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.show()
```

```python
# 17. Term Deposit Subscriptions
plt.figure(figsize=(6, 4))
data['y'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'salmon'])
plt.title('Term Deposit Subscriptions')
plt.ylabel('')
plt.show()
```

```python
# 18. Correlation Analysis

numeric_data = data.select_dtypes(include=['number'])


plt.figure(figsize=(12, 8))
sns.heatmap(numeric_data.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap of Numeric Attributes')
plt.show()

```
